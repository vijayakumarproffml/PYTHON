```python
# centrality 
# rendu graph ku idaila oru node bridge ah act aagum

import networkx as nx
import matplotlib.pyplot as plt

G1=nx.complete_graph(5)
G2=nx.complete_graph(5)
G2=nx.relablel_nodes(G2,{0:"A",1:'B',2:'C',3:'D',4:'E'})

G_connecter=nx.from_edgelist([(4,'X'),('X','A')])

G=nx.compose_all([G1,G2,G_connecter])

print(nx.degree_centrality(G))
print(nx.betweenness_centrality(G))

print(nx.density(G)) #
print(nx.diameter(G)) # longest shortest path

#Euler, intha case la all nodes ah one time matume visit pannum
print(list(nx.eulerian_path(G))) 
print(list(nx.find_cliques(G)))


nx.draw_spring(G,with_labels=True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6)])

# Breadth-First Search (BFS)
def bfs(graph, start):
    visited = set()
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.add(node)
            print(node, end=' ')
            neighbors = graph.neighbors(node)
            queue.extend(neighbors)

            # Visualization: Highlight the visited node
            nx.draw(graph, pos=nx.spring_layout(graph), with_labels=True, node_color='skyblue', node_size=700)
            nx.draw_networkx_nodes(graph, pos=nx.spring_layout(graph), nodelist=[node], node_color='red', node_size=700)
            plt.title('BFS Traversal')
            plt.pause(1)
            plt.clf()

print("BFS Traversal:")
bfs(G, 1)  # Start BFS from node 1

# Depth-First Search (DFS)
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')

    for neighbor in graph.neighbors(start):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

            # Visualization: Highlight the visited node
            nx.draw(graph, pos=nx.spring_layout(graph), with_labels=True, node_color='skyblue', node_size=700)
            nx.draw_networkx_nodes(graph, pos=nx.spring_layout(graph), nodelist=[neighbor], node_color='red', node_size=700)
            plt.title('DFS Traversal')
            plt.pause(1)
            plt.clf()

print("\n\nDFS Traversal:")
dfs(G, 1)  # Start DFS from node 1
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to add edges to the graph
def add_edge(graph, u, v):
    graph.add_edge(u, v)
    graph.add_edge(v, u)

# Breadth-First Search algorithm
def bfs(graph, start):
    visited = set()
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            queue.extend(graph.neighbors(node))
            
            

# Depth-First Search algorithm
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=" ")
    for neighbor in graph.neighbors(start):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Visualization function
def visualize_graph(graph):
    nx.draw(graph, with_labels=True)
    plt.show()

# Creating a graph
G = nx.Graph()
add_edge(G, 0, 1)
add_edge(G, 0, 2)
add_edge(G, 1, 3)
add_edge(G, 1, 4)
add_edge(G, 2, 5)
add_edge(G, 2, 6)

# Visualizing the graph
visualize_graph(G)

# Performing BFS starting from node 0
print("BFS:")
bfs(G, 0)
print()

# Performing DFS starting from node 0
print("DFS:")
dfs(G, 0)
print()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
# G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6)])
G.add_edges_from([(1,2),(2,3),(3,4),(3,5),(4,6),(6,7),(2,8),(8,9),(9,4)])

# Breadth-First Search (BFS)
def bfs(graph, start):
    visited = set()
    queue = [start]
    pos = nx.spring_layout(graph)  # Fixed position of nodes

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.add(node)
            print(node, end=' ')
            neighbors = graph.neighbors(node)
            queue.extend(neighbors)

            # Visualization: Highlight the visited node
            nx.draw(graph, pos=pos, with_labels=True, node_color=['red' if n == node else 'skyblue' for n in graph.nodes()], node_size=700)
            plt.title('BFS Traversal')
            plt.pause(1)
            plt.clf()

# Depth-First Search (DFS)
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')
    pos = nx.spring_layout(graph)  # Fixed position of nodes

    for neighbor in graph.neighbors(start):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

            # Visualization: Highlight the visited node
            nx.draw(graph, pos=pos, with_labels=True, node_color=['red' if n == neighbor else 'skyblue' for n in graph.nodes()], node_size=700)
            plt.title('DFS Traversal')
            plt.pause(1)
            plt.clf()

print("BFS Traversal:")
bfs(G, 1)  # Start BFS from node 1

print("\n\nDFS Traversal:")
dfs(G, 1)  # Start DFS from node 1
```
