```python
# create a Graph

class GRAPH:
    # creating a empy Graph
    
    def __init__(self):
        self.graph={}
        
    # add vertex
    
    def AddVertex(self,vertex):
        if vertex not in self.graph:
            self.graph[vertex]=[]
            
        else: # inga else podamal irupathu nallathu, result la multiple times varum
            print("Vertex Already Exists")
            
    # add Edge
    
    def AddEdge(self,vertex1,vertex2,isDirected=False):
        self.AddVertex(vertex1)
        self.AddVertex(vertex2)
        
        self.graph[vertex1].append(vertex2)
        if not isDirected:
            self.graph[vertex2].append(vertex1)
            
    # Display
    
    def Display(self):
        for keys, values in self.graph.items():
            print(f"{keys} => {values}")
            
            
    # vertex only
    
    def Get_vertices(self):
         for i in self.graph:
             print("Get Vertices:", i)
             
             
    # Edges only
            
    def Get_Edges(self):
        print("GetEdges:")
        
        for keys, values in self.graph.items(): # inga error vanthathu because items ku bathila item nu pottathal
            
            for adjacent_list_values in values:
                
                print(f"({keys},{adjacent_list_values})")
                
                
    # remove vertex
    
    def Removevertex(self,vertex):
        #vertex only {'B':['A','C','D']}  ithu full ah remove aagirum
        if vertex in self.graph:
            del self.graph[vertex]
            
        # values la iruka vertex ah remove panrathuku
        for keys,values in self.graph.item():
            if vertex in values:
                values.remove(vertex)
                
    #remove edges
    def isEdge(self,vertex1,vertex2):
        return vertex1 in self.graph[vertex2] or vertex2 in self.graph[vertex1]
    
    def RemoveEdge(self,vertex1,vertex2):
        print(" After Remove Edge")
        if self.isEdge(vertex1,vertex2): # self podakoodathu error varum(positional argument error)
            self.graph[vertex1].remove(vertex2)
            self.graph[vertex2].remove(vertex1)
        else:
            print(f"No edge exists between {vertex1} and {vertex2}.")
            
    
            
graph1=GRAPH()
graph1.AddEdge('A','B')
graph1.AddEdge('B','C')
graph1.AddEdge('B','D')
graph1.AddEdge('C','D')
graph1.Display()
graph1.Get_vertices()
graph1.Get_Edges()
# graph1.Removevertex('B')            
graph1.RemoveEdge('A','B')
graph1.Display()
           
        
        
         
                
    
        

        
            
            
        
        
        
```
