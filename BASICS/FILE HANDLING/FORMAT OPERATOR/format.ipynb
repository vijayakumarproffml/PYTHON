```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def build_expression_tree(infix):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
    stack = []
    output = []

    for token in infix:
        if token.isalnum():  # Operand
            output.append(TreeNode(token))
        elif token == '(':
            stack.append(token)
        elif token == ')':
            while stack and stack[-1] != '(':
                output.append(TreeNode(stack.pop()))
            stack.pop()  # Discard '('
        else:  # Operator
            while stack and stack[-1] != '(' and precedence[stack[-1]] >= precedence[token]:
                output.append(TreeNode(stack.pop()))
            stack.append(token)

    while stack:
        output.append(TreeNode(stack.pop()))

    for token in output:
        if token.value in precedence:  # Operator
            token.right = stack.pop()
            token.left = stack.pop()
        stack.append(token)

    return stack[0]

def print_tree(node, level=0):
    if node is not None:
        print("  " * level + str(node.value))
        print_tree(node.left, level + 1)
        print_tree(node.right, level + 1)

# Example usage:
infix_expression = ['a' ,'b' ,'c', '*' ,'d' ,'e' ,'f' ,'*','+', 'g' ,'-', 'h', 'i', 'j', '*' ,'+' ,'k', 'l' ,'m', '+' ,'n', 'o' ,'*','+' ,'/', '-']
root = build_expression_tree(infix_expression)
print("Expression Tree:")
print_tree(root)
```
